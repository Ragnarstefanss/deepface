## Verify if two images are of the same person

```python
def verify(img1_path, img2_path, model_name = "VGG-Face"):
    img1 = cv2.imread(img1_path)
    img2 = cv2.imread(img2_path)
    
    plt.imshow(img1[:, :, ::-1])
    plt.show()
    plt.imshow(img2[:, :, ::-1])
    plt.show()
    
    result = DeepFace.verify(img1_path, img2_path, model_name = model_name)
    #print("results:", result)
    
    verification = result["verified"]
    if verification:
        print("both images are of the same person")
    else:
        print("images are not of the same person")
        
    #return verification
```

```python
def verify_same_person(image_path1, image_path2):
    return DeepFace.verify(image_path1, image_path2, detector_backend = backends[0])

def find_related_images(image_path, enforce_detection=True):
    return DeepFace.find(img_path = image_path, db_path = "database", detector_backend = backends[0], enforce_detection=enforce_detection)

def analyse_emotion(image_path):
    return DeepFace.analyze(image_path, detector_backend = backends[0])

# show images
def show_single_image(image_path):
    img1 = cv2.imread(image_path)
    plt.imshow(img1[:, :, ::-1])
    plt.show()
    
def show_related_images(image_path):
    df = find_related_images(image_path, False)
    for item in df["identity"]:
        show_single_image(item)
```
